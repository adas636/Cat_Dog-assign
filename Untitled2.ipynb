```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import os
from pathlib import Path
import librosa
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.model_selection import cross_validate, StratifiedKFold
from sklearn.metrics import make_scorer, f1_score, recall_score, precision_score
import optuna
```

```python
def extract_features(directory):
    features = []
    labels = []
    for folder in os.listdir(directory):
        for filename in os.listdir(directory+str(folder)):
            f = os.path.join(directory+str(folder), filename)
            if os.path.isfile(f):
                x, sr = librosa.load(f, res_type='kaiser_fast',sr=None)
                mfccs = np.mean(librosa.feature.mfcc(x,sr=sr,n_mfcc=100).T,axis=0)
                features.append(mfccs)
                labels.append(f.split('\\')[-2])
    return [labels, features]
```

```python
print(str(Path.cwd()) + '\\train\\')
labels_train, features_train = extract_features(str(Path.cwd()) + '\\train\\')
labels_test, features_test = extract_features(str(Path.cwd()) + '\\test\\')
```

```python
print(labels_train)
print(labels_test)
```

```python
X_train = np.array(features_train)
X_test = np.array(features_test)
le = LabelEncoder()
Y_train = le.fit_transform(labels_train)
Y_test = le.fit_transform(labels_test)
```

```python
n_neighbors = 8
cls = KNeighborsClassifier(n_neighbors)
cls.fit(X_train, Y_train)
Y_predict = cls.predict(X_test)
print(Y_test.shape)
print(Y_predict.shape)
```

```python
print(f"Accuracy:{metrics.accuracy_score(Y_test, Y_predict)}\nF_1 score:{f1_score(Y_test, Y_predict)}\nRecall score:{recall_score(Y_test, Y_predict)}\nPrecision score:{precision_score(Y_test, Y_predict)}")         
```

```python
model = KNeighborsClassifier

scoring = {'f1_macro': make_scorer(f1_score, average='macro')}

def get_space(trial):
    space ={
        "n_neighbors": trial.suggest_int("n_neighbors", 3, 20),
        "algorithm": trial.suggest_categorical('algorithm', ["auto", "ball_tree", "kd_tree", "brute"]),
        "leaf_size": trial.suggest_int("leaf_size", 1, 50),
        "n_jobs": trial.suggest_int("n_jobs", -1, -1),
        "weights": trial.suggest_categorical("weights", ["uniform", "distance"])
    }
    return space

trials = 60

def objective(trial, model, X, y):
    model_space = get_space(trial)
    mdl = model(**model_space)
    scores = cross_validate(mdl, X, y, scoring=scoring, cv=StratifiedKFold(n_splits=5), return_train_score=True)
    return np.mean(scores['test_f1_macro'])
```

```python
study = optuna.create_study(direction='maximize')
```

```python
%%time
study.optimize(lambda x: objective(x, model, X_train, Y_train), n_trials=trials, gc_after_trial=True, show_progress_bar=True)
print('params: ', study.best_params)
```

```python
cls_2 = KNeighborsClassifier(n_neighbors=study.best_params['n_neighbors'], algorithm=study.best_params['algorithm'], leaf_size=study.best_params['leaf_size'], n_jobs=-1, weights=study.best_params['weights'])
cls_2.fit(X_train, Y_train)
Y_2_predict = cls_2.predict(X_test)
print(f"Accuracy:{metrics.accuracy_score(Y_test, Y_2_predict)}\nF_1 score:{f1_score(Y_test, Y_2_predict)}\nRecall score:{recall_score(Y_test, Y_2_predict)}\nPrecision score:{precision_score(Y_test, Y_2_predict)}")
```

```python
cls_3 = RandomForestClassifier(random_state=42)
cls_3.fit(X_train, Y_train)
Y_3_predict = cls_3.predict(X_test)
print(f"Accuracy:{metrics.accuracy_score(Y_test, Y_3_predict)}\nF_1 score:{f1_score(Y_test, Y_3_predict)}\nRecall score:{recall_score(Y_test, Y_3_predict)}\nPrecision score:{precision_score(Y_test, Y_3_predict)}")
```
